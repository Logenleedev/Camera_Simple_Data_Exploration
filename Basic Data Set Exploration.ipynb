### Plot a Zoom wide bar plot matplotlib?

```python
facet = sns.FacetGrid(data,aspect=4)
facet.map(sns.kdeplot,'Zoom wide (W)',shade= True)
facet.set(xlim=(0, data['Zoom wide (W)'].max()))
facet.add_legend()
```
